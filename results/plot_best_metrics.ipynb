```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
# plt.rcdefaults()
```

```python
df=pd.read_csv("best_performance_metrics.csv")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["Dimensions"], df["Accuracy"], label="Accuracy")
plt.plot(df["Dimensions"], df["F1"], label="F1 Score")
plt.plot(df["Dimensions"], df["KLL"], label="KLL")
plt.plot(df["Dimensions"], df["RTE"], label="RTE")
plt.xlabel("Matrix Dimension")
plt.ylabel("Metric Value")
plt.title("Metrics vs Matrix Dimension")
plt.legend()
plt.grid(True)
plt.show()
```
